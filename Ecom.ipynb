Converting awbno and SKU to string

```python
oms['SKU']=oms['SKU'].astype("string")
oms['awbno']=oms['awbno'].astype("string")
pf['SKU']=pf['SKU'].astype("string")
pf['awbno']=pf['awbno'].astype("string")
```

```python
oms['SKU']=oms['SKU'].str.strip()
oms['awbno']=oms['awbno'].str.strip()
pf['SKU']=pf['SKU'].str.strip()
pf['awbno']=pf['awbno'].str.strip()
```

```python
print(oms.dtypes['awbno'])
print(oms.dtypes['SKU'])
print(pf.dtypes['awbno'])
print(pf.dtypes['SKU'])
```

Dropping "Unnamed: 0" column

```python
oms.drop('Unnamed: 0',axis=1,inplace=True)
pf.drop('Unnamed: 0',axis=1,inplace=True)
```

Dropping Duplicates from oms and pf





```python
oms.drop_duplicates(inplace=True)
pf.drop_duplicates(inplace=True)
```

```python
oms.shape
```

```python
pf.shape
```

Merging sku and awbno for both datasets

```python
oms['SK_aws']=oms['SKU']+"-"+oms['awbno']
pf['SK_aws']=pf['SKU']+"-"+pf['awbno']
```

```python
oms.head()
```

```python
oms.isna().sum()
pf.isna().sum()
```

```python
pf.head()
```

Merging oms and pf

```python
mg_df=oms.merge(pf,on='SK_aws',how='left',suffixes=('_oms','_pf'))
```

```python
mg_df.head()
```

```python
mg_df.shape
```

```python
mg_df.isna().sum()
```

Replacing Null values with Blank

```python
mg_df['awbno_pf']=mg_df['awbno_pf'].fillna('')
mg_df['awbno_oms']=mg_df['awbno_oms'].fillna('')
mg_df['SKU_pf']=mg_df['SKU_pf'].fillna('')
mg_df['inventory_status']=mg_df['inventory_status'].fillna('')
mg_df['getStatus']=mg_df['getStatus'].fillna('')
```

```python
mg_df.isna().sum()
```

Logic for "final_inventory_status"

```python
def get_inventory_status(row):
    # Check if both awbno and sku match
    if row['awbno_oms'] == row['awbno_pf'] and row['SKU_oms'] == row['SKU_pf'] :
        return row['getStatus']
    elif row['awbno_oms'] == row['awbno_pf']:
                return str(row['getStatus']) + "_sku-mismatch"
    else:
                return "missing"
mg_df['final_inventory_status'] = mg_df.apply(get_inventory_status, axis=1)

# Display the result
```

Converting Merged Data into a CSV file

```python
mg_df.to_csv('final_df.csv')
```
